```python
import pandas as pd

def percent_counts(df, column_name):
    # Calculate the percentage counts of unique values in the specified column
    value_counts = df[column_name].value_counts(normalize=True) * 100

    # Create a DataFrame to store the values and their percentages
    value_counts_df = pd.DataFrame({column_name: value_counts.index, 'Percentage': value_counts.values})

    # Apply a background gradient style to the DataFrame using the 'coolwarm' colormap
    styled_df = value_counts_df.style.background_gradient(cmap='coolwarm').set_precision(2)

    # Return the styled DataFrame
    return styled_df
spec_columns2 = ['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope', 'HeartDisease']
for column in spec_columns2:
    result = percent_counts(df, column)
    print(f"Column: {column}")
    display(result)
    print("\n")
```

```python
X = heart.drop(['HeartDisease'],axis=1) # to remove the HeartDisease columns
y = heart['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=21)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
std_scaler = StandardScaler()

# Fit the scaler to your feature data (X)
std_scaler.fit(X)

# Transform the training and test data using the fitted scaler
X_train_std = std_scaler.transform(X_train)
X_test_std = std_scaler.transform(X_test)

# Create DataFrames for the standardized training and test data
X_train = pd.DataFrame(X_train_std, columns=X.columns)
X_test = pd.DataFrame(X_test_std, columns=X.columns)
from IPython.display import display

# Display the first few rows of X_train
display(X_train.head())

# Display the first few rows of X_test
display(X_test.head())
```

Model Training and Find the Accuracy of the model using Logistic Regression Models

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
lgreg = LogisticRegression()
lgreg.fit(X_train, y_train)
y_pred_log = lgreg.predict(X_test)
```

```python
lg_train = round(lgreg.score(X_train, y_train) * 100, 2)
lg_accuracy = round(accuracy_score(y_pred_log, y_test) * 100, 2)
lg_f1 = round(f1_score(y_pred_log, y_test) * 100, 2)
```

```python
print("Training Accuracy    :",lg_train,"%")
print("Model Accuracy Score :",lg_accuracy,"%")
print("Confusion Matrix: \n",confusion_matrix(y_test,y_pred_log))
print("Classification_Report: \n",classification_report(y_test,y_pred_log))
```

Using Logistic Regression accuracy of model is 92%(appro.)

Model Training and Find the Accuracy of the model using DecisionTreeClassifier Models

```python
# Create and train the model
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
y_pred_dt = dt_classifier.predict(X_test)

# Calculate evaluation metrics
dt_train = round(dt_classifier.score(X_train, y_train) * 100, 2)
dt_accuracy = round(accuracy_score(y_pred_dt, y_test) * 100, 2)
dt_f1 = round(f1_score(y_pred_dt, y_test) * 100, 2)

# Print results
print("Training Accuracy    :", dt_train, "%")
print("Model Accuracy Score :", dt_accuracy, "%")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))
print("Classification Report:\n", classification_report(y_test, y_pred_dt))
```

Using DecisionTreeClassifier accuracy of model is 88%(appro.)

Model Training and Find the Accuracy of the model using SVC Models

```python
svm_classifier = SVC()
svm_classifier.fit(X_train, y_train)
y_pred_svm = svm_classifier.predict(X_test)

svm_train = round(svm_classifier.score(X_train, y_train) * 100, 2)
svm_accuracy = round(accuracy_score(y_pred_svm, y_test) * 100, 2)
svm_f1 = round(f1_score(y_pred_svm, y_test) * 100, 2)

print("Training Accuracy    :", svm_train, "%")
print("Model Accuracy Score :", svm_accuracy, "%")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))
print("Classification Report:\n", classification_report(y_test, y_pred_svm))
```

```python
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)
y_pred_rf = rf_classifier.predict(X_test)

rf_train = round(rf_classifier.score(X_train, y_train) * 100, 2)
rf_accuracy = round(accuracy_score(y_pred_rf, y_test) * 100, 2)
rf_f1 = round(f1_score(y_pred_rf, y_test) * 100, 2)

print("Training Accuracy    :", rf_train, "%")
print("Model Accuracy Score :", rf_accuracy, "%")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
```
